```python
from pyproj import Proj, transform

def mercator_projection(lon, lat):
    in_proj = Proj('epsg:4326')  # WGS84 Latitude/Longitude
    out_proj = Proj('ESRI:53030')  # Projection Robinson

    # Fonction pour convertir lat/long en x/y
    def project(lat, lon):
        x, y = transform(in_proj, out_proj, lon, lat)
        return x, y
# Example usage:
longitude = 12.0
latitude = 13.0
print(mercator_projection(longitude, latitude))
```
