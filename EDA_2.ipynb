```python
import math

def lat_lng_to_pixel(latitude, longitude, image_width, image_height):
    # Convertir la latitude en radians
    lat_rad = math.radians(latitude)
    
    # Calculer la projection de Mercator
    x = (longitude + 180) * (image_width / 360)
    y = (image_height / 2) - (image_width * math.log(math.tan((math.pi / 4) + (lat_rad / 2))) / (2 * math.pi))
    
    return x, y

lat_lng_to_pixel(48, 2.3, 2000, 857)
```
