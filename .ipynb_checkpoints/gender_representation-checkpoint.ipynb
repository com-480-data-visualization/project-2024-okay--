```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.widgets import Slider

# Load the datasets
athletes = pd.read_csv('data/dataset_olympics.csv')
noc_regions = pd.read_csv('data/noc_region.csv')
gdf_world = gpd.read_file('data/countries.geojson')

# Create a dictionary from noc_regions DataFrame
noc_to_country = pd.Series(noc_regions.reg.values, index=noc_regions.noc_region).to_dict()

# Helper function to map NOC to country names and setup the map
def setup_map(year):
    # Filter data for the selected year
    year_athletes = athletes[athletes['Year'] == year]
    unique_athletes = year_athletes.drop_duplicates(subset=['ID', 'NOC', 'Sex'])
    grouped_athletes = unique_athletes.groupby(['NOC', 'Sex']).size().unstack(fill_value=0)

    # Ensure both 'male_count' and 'female_count' exist
    if 'M' not in grouped_athletes.columns:
        grouped_athletes['M'] = 0
    if 'F' not in grouped_athletes.columns:
        grouped_athletes['F'] = 0

    grouped_athletes.rename(columns={'M': 'male_count', 'F': 'female_count'}, inplace=True)
    grouped_athletes['female_male_ratio'] = grouped_athletes['female_count'] / (grouped_athletes['male_count'] + grouped_athletes['female_count'])
    grouped_athletes['country'] = grouped_athletes.index.map(lambda noc: noc_to_country.get(noc, 'Unknown'))

    merged_gdf = gdf_world.merge(grouped_athletes, how='left', left_on='ADMIN', right_on='country')
    
    ax.clear()
    ax.set_axis_off()
    merged_gdf.plot(column='female_male_ratio', ax=ax, cmap=plt.cm.viridis, norm=norm, legend=True, missing_kwds={'color': 'lightgrey'})

# Initialize plot settings
fig, ax = plt.subplots(figsize=(15, 10))
plt.subplots_adjust(bottom=0.25)
norm = mcolors.Normalize(vmin=0, vmax=1)

# Slider
years = sorted(athletes['Year'].unique())
ax_slider = plt.axes([0.2, 0.1, 0.6, 0.03])
slider = Slider(ax_slider, 'Year', valmin=min(years), valmax=max(years), valinit=min(years), valstep=4)

# Update the plot based on the slider
def update(val):
    setup_map(int(val))
    fig.canvas.draw_idle()

slider.on_changed(update)

# Initial map setup
setup_map(min(years))

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.widgets import Slider

# Load the datasets
athletes = pd.read_csv('data/dataset_olympics.csv')
noc_regions = pd.read_csv('data/noc_region.csv')
gdf_world = gpd.read_file('data/countries.geojson')

# Create a dictionary from noc_regions DataFrame
noc_to_country = pd.Series(noc_regions.reg.values,index=noc_regions.noc_region).to_dict()

# Helper function to map NOC to country names
def map_noc_to_country(noc):
    return noc_to_country.get(noc, None)

# Calculate the number of unique female and male athletes per country
unique_athletes = athletes.drop_duplicates(subset=['ID', 'NOC', 'Sex'])
grouped_athletes = unique_athletes.groupby(['NOC', 'Sex']).size().unstack().fillna(0)

# Rename columns to 'male_count' and 'female_count'
grouped_athletes.rename(columns={'M': 'male_count', 'F': 'female_count'}, inplace=True)

# Calculate the female to male ratio
grouped_athletes['female_male_ratio'] = grouped_athletes['female_count'] / (grouped_athletes['male_count'] + grouped_athletes['female_count'])

# Map NOC to country names
grouped_athletes['country'] = grouped_athletes.index.map(map_noc_to_country)

# Merge with GeoDataFrame
merged_gdf = gdf_world.merge(grouped_athletes, how='left', left_on='ADMIN', right_on='country')

cmap = plt.cm.viridis

# Plotting
fig, ax = plt.subplots(1, figsize=(25, 40))
ax.set_axis_off()

# Normalize the color scale to the range of the ratio
norm = mcolors.Normalize(vmin=0, vmax=1)

# Plot the countries with valid 'female_male_ratio' values
merged_gdf.dropna(subset=['female_male_ratio']).plot(column='female_male_ratio', ax=ax, legend=False, cmap=plt.cm.viridis, norm=norm)

# Plot countries with NaN values in 'female_male_ratio' in white to show missing data
merged_gdf[merged_gdf['female_male_ratio'].isna()].plot(ax=ax, color='white')

# Colorbar
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=2.5)
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=plt.cm.viridis), cax=cax)

# Annotate the colorbar
cax.text(4.4, -0.02, 'No Female Athletes', va='bottom', ha='right', transform=cax.transAxes)
cax.text(6, 1.01, 'Equal Female and Male Athletes', va='top', ha='right', transform=cax.transAxes)

ax_slider = divider.append_axes("bottom", size="2%", pad=0.5)
slider = Slider(ax_slider, 'Slider', valmin=0, valmax=1, valinit=0.5)

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pycountry
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Load the dataset
athletes = pd.read_csv('data/dataset_olympics.csv')
gdf_world = gpd.read_file('data/countries.geojson')

# Calculate the number of female and male athletes per country
athletes['count'] = 1
grouped_athletes = athletes.groupby(['NOC', 'Sex']).count()['count'].unstack().fillna(0)

# Ensure the columns are named correctly according to the Sex values "M" and "F"
grouped_athletes.columns = ['F', 'M']

# Calculate the female to male ratio, using a small epsilon to avoid division by zero
epsilon = 1e-6
grouped_athletes['female_male_ratio'] = grouped_athletes['F'] / (grouped_athletes['M'] + epsilon)

# Map country codes (NOC) to country names using pycountry
def map_country_codes(code):
    try:
        return pycountry.countries.lookup(code).name
    except LookupError:
        return None

# Map NOC to country names
grouped_athletes['country'] = grouped_athletes.index.map(map_country_codes)
```

```python
# Merge with GeoDataFrame
merged_gdf = gdf_world.merge(grouped_athletes, how='left', left_on='ADMIN', right_on='country')

# Define single color gradient: shades of purple
cmap = mcolors.LinearSegmentedColormap.from_list("ratio", ["#e0bbe4", "#8e44ad"], N=256)

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.1)

# Normalize the color scale to the range of the data
norm = mcolors.Normalize(vmin=merged_gdf['female_male_ratio'].min(), vmax=merged_gdf['female_male_ratio'].max())

# Plot the countries with valid 'female_male_ratio' values
merged_gdf.dropna(subset=['female_male_ratio']).plot(column='female_male_ratio', ax=ax, legend=True, cmap=cmap, norm=norm, cax=cax, legend_kwds={'orientation': "horizontal"})

# Plot countries with NaN values in 'female_male_ratio' in white to show missing data
merged_gdf[merged_gdf['female_male_ratio'].isna()].plot(ax=ax, color='white')

plt.show()
```
